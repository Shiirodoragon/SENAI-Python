## Calculo IMC

Crie um programa que faça o Calculo do IMC das pessoas onde peça as seguintes informações:

- Altura em metros

- Peso com casa decimal

- Pergunte a idade da pessoa

utilize as função que vocês aprenderam para caso o paciente coloque virgula, e siga a tabela a seguir para mostrar a informação para nosso usuário

| IMC | CLASSIFICAÇÃO |
| --- | ------------- |
| < 18,5 | MAGREZA |
| ENTRE 18,5 E 24,9 | NORMAL |
| ENTRE 25 E 29,9 | SOBREPESO |
| ENTRE 30 E 34,9 | OBESIDADE GRAU 1 |
| ENTRE 35 E 39,9 | OBESIDADE GRAU 2 |
| ACIMA DE 40 | OBESIDADE GRAU 3 |


`Para fazer o calculo` = peso / altura²

Caso o paciente tenha menos que 15 anos e esteja com a classificação maior que 30 mostre que está na obesidade infantil

Além disso utilizem a seguinte tabela para mostrar de uma forma visual cada classificação, seguindo as seguites cores:

| COR | CLASSIFICAÇÃO |
| --- | ------------- |
| CINZA | MAGREZA |
| VERDE | NORMAL |
| PRETO | SOBREPESO |
| MAGENTA | OBESIDADE GRAU 1 |
| CIANO | OBESIDADE GRAU 2 |
| AMARELO | OBESIDADE GRAU 3 |
| VERMELHO EM NEGRITO | OBESIDADE INFANTIL |

No final do programa mostre como foi feito o calculo para chegar em tal resultado e mostre o resultado

```python
altura = float(input("Insira sua altura em METROS  ").strip().replace(",","."))
peso = float(input("Insira seu peso  ").strip().replace(",","."))
idade = int(input("Insira sua idade  ").strip())

if(altura<=100):
  altura /= 100

imc = peso/(altura**2)

if(idade <= 15 and imc >= 30):
  print(f"\033[1;31mOBESIDADE INFANTIL")

elif(imc < 18.5):
  print(f"\033[0;37mMAGREZA")

elif(18.5 <= imc <= 24.9):
  print(f"\033[0;32mNORMAL")

elif(24.9 < imc <= 29.9):
  print(f"\033[0mSOPREPESO")

elif(29.9 < imc <= 34.9):
  print(f"\033[0;35mOBESIDADE GRAU 1")

elif(34.9 < imc <= 39.9):
  print(f"\033[0;36mOBESIDADE GRAU 2")

else:
  print(f"\033[0;33mOBESIDADE GRAU 3")

print(f"\033[0mCálculo: {peso}/{altura}^2 = {imc:.1f}")
```
