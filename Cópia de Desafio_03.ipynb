# Desafio `for`

Crie um programa onde a pessoa escreva um texto que possa receber tanto caracteres quanto números e após isso ele informe para a pessoa:



1.   Quantas letras possui
2.   Quantos digitos possui



Link para entender como verificar qual é o caracter: https://www.w3schools.com/python/python_ref_string.asp

```python
frase = str(input("Insira sua frase  ").strip())
digito = 0
letra = 0
for i in str(frase):

  if(i.isnumeric()):
    digito += 1

  elif(i.isalpha()):
    letra += 1

print(f"{digito} dígitos, {letra} letras")
```
