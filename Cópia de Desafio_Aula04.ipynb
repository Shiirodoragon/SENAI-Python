# Desafio `while`

Crie um programa onde você é dono de um mercadinho e você possui a tabela com os produtos e o preço de cada produto, quando o cliente for entrar no mercadinho solicite o nome dele para cadastro e gere um id aleatório para ele, após isso conforme ele for solicitando você vai adicionando os produtos no carrinho dele, e quando ele finalizar mostre para ele o valor da compra.

1 - Seu mercadinho tem que ter um nome

2 - tem que aparecer uma lista com as opções, o código e valor de cada opção

3 - Abaixo segue a planilha com o código, o nome do produto e o valor do produto

image.png

4 - mostrar o valor total para ele

5 - dar 4 opções de pagamento que serão (dinheiro, pix, débito ou crédito)

6 - aplicar o desconto seguindo a planilha abaixo e mostre para o usuário o valor original e o valor com desconto

7 - No final de tudo mostre o valor que será pago e utilize uma biblioteca para verificar o horário e dizer
  - menor 12 = tenha um excelente dia
  - menor 19 = tenha uma excelente tarde
  - maior 19 = tenha uma excelente noite

```python
import random
import datetime

print(f"Boas vindas ao Mercadinho Muchiba, vou precisar de uns dados para fazer o seu cadastro, tá bom? :)")
nome = str(input("Primeiro eu preciso do seu nome  ").strip().title())
id = random.randrange(1000,9999)
valor = 0
print(f"Legal {nome}, seu id de cliente é {id}")

print(f"""Aqui está nosso catálogo de produtos
101 Banana R$5,70
102 Maçã R$8,80
103 Batata R$5,30
104 Alface R$3,50
510 Macarrão R$4,00
520 Feijão R$3,00
530 Arroz R$5,80
235 Salgadinho R$2,00
236 Bolacha R$1,50 """)
produto = abs(int(input("Digite o código do produto que deseja comprar!  ").strip()))

if produto == 101:
  valor += 5.7
elif produto == 102:
  valor += 8.8
elif produto == 103:
  valor += 5.3
elif produto == 104:
  valor += 3.5
elif produto == 520:
  valor += 3
elif produto == 530:
  valor += 5.8
elif produto == 235:
  valor += 2
elif produto == 236:
  valor += 1.5
else:
  print("Não entendi, insira de novo o número do produto  ")

cont = input(f"Legal! vou colocar o produto {produto} no seu carrinho! Mais alguma coisa? (S/N)")

while cont == 'S':
  produto = abs(int
   (input("Digite o código do produto que deseja comprar!  ").strip()))

  if produto == 101:
    valor += 5.7
  elif produto == 102:
    valor += 8.8
  elif produto == 103:
    valor += 5.3
  elif produto == 104:
    valor += 3.5
  elif produto == 520:
    valor += 3
  elif produto == 530:
    valor += 5.8
  elif produto == 235:
    valor += 2
  elif produto == 236:
    valor += 1.5
  else:
    print("Não entendi, insira de novo o número do produto  ")

  cont = input(f"Legal! vou colocar o produto {produto} no seu carrinho! Mais alguma coisa? (S/N)")

if cont == "N":
  pag = abs(int(input(f"""A sua compra está totalizando R${valor:.2f}. Como você gostaria de pagar?
  1. Dinheiro
  2. PIX
  3. Débito
  4. Crédito  """).strip()))

  if pag == 1:
    valor *= 0.1
  elif pag == 2:
    valor *= 0.2
  elif pag == 3:
    valor *= 0.03
  elif pag == 4:
    valor *= 1.25
  else:
    print("Opa, não entendi! Você pode digitar de novo?")
    pag = abs(int(input(f"""A sua compra está totalizando R${valor:.2f}. Como você gostaria de pagar?
  1. Dinheiro
  2. PIX
  3. Débito
  4. Crédito  """)).strip())

  print(f"Depois das devidas taxas e descontos, o seu total é R${valor:.2f}!")
  print(f"Obrigado por comprar no Mercadinho Muchiba!", end = " ")

  hora = datetime.time().hour
  if hora < 12:
    print("Tenha um excelente dia! :)")
  elif 12 <= hora < 19:
    print("Tenha uma excelente tarde! :)")
  else:
    print("Tenha uma excelente noite!")
```
