# Soma Par e Impar

Crie um programa que o usuario pode digitar quantos números desejar, no final do programa mostre quantos números par ele digitou, quantos números impar ele digitou, a soma dos valores do números pares, a soma dos valores dos números impares

```python
num = int(input("Insira um número  ").strip().replace(",","."))
cont = str(input("Deseja continuar? (S/N)  ").strip().upper())
par = 0
impar = 0
p = 0
i = 0
if num%2 == 0:
  par += num
  p += 1

else:
  impar += num
  i += 1

while cont == "S":
  num = int(input("Insira um número  ").strip().replace(",","."))
  cont = str(input("Deseja continuar? (S/N)  ").strip().upper())

  if num%2 == 0:
    par += num
    p += 1

  else:
    impar += num
    i += 1

print(f"Você digitou {i} números ímpares, cuja soma é igual a {impar}, e {p} números pares, cuja soma é igual a {par}")
```

# Cadastro Senha

Crie um programa onde o nosso usuario vai cadastrar uma senha mas ela tem que ter alguns requisitos

- Ter no mínimo 8 caracteres

- Ter no minimo 1 letra maiuscula

- Ter no minimo 1 letra minuscula

- Ter no minimo 1 número

- Ter no minimo 1 caracter especial

```python
senha = str(input("Insira sua senha:  "))
up = 0
low = 0
num = 0
espc = 0
x= True

for c in senha:
  if len(senha) < 8:
    x = False

  if c.isupper():
    up += 1
    if up == 0:
      x = False

  if c.islower():
    low += 1
    if low == 0:
      x = False

  if c.isnumeric():
    num += 1
    if num == 0:
      x= False

  if not (c.isalpha() and c.isspace() and c == "_"):
    espc += 1
    if espc == 0:
      x = False


while not x:
  print(f'''Sua senha não é válida, tente novamente. A senha deve:
Ter no mínimo 8 caracteres
Ter no minimo 1 letra maiuscula
Ter no minimo 1 letra minuscula
Ter no minimo 1 número
Ter no minimo 1 caracter especial''')
  senha = str(input("Insira sua senha:  "))


print(f"up = {up}; low = {low}, num = {num}, espc = {espc}")
print(f"Sua senha é válida")
```

# Temperatura

Faça um programa em linguagem Python utilizando while que recebe a temperatura de z cliente e imprima a mensagem de se a temperatura esta normal (menor que 37,2 C) ou está em estado febril (37,3 C a 38 C ) ou com febre (38 C a 39 C) e com febre alta(acima 39 C). No final mostre a quantidade de pessoas analisadas e a média de temperatura.

```python
temp = float(input("Insira a temperatura medida  ").strip().replace(",","."))
total = temp
p = 1
if temp < 37.2:
  print("Temperatura normal")

elif 37.2 <= temp < 38:
  print("Estado febril")

elif 38 <= temp < 39:
  print("Febre")

else:
  print("Febre alta")

cont = str(input("Mais algum paciente?  (S/N)  ").strip().upper())

while cont == "S":
  temp = float(input("Insira a temperatura medida  ").strip().replace(",","."))
  if temp < 37.2:
    print("Temperatura normal")

  elif 37.2 <= temp < 38:
    print("Estado febril")

  elif 38 <= temp < 39:
    print("Febre")

  else:
    print("Febre alta")

  cont = str(input("Mais algum paciente?  (S/N)  ").strip().upper())

  total += temp
  p += 1

media = total/p
print(f"Analisei {p} pessoa(s) e a média de temperatura é {media:.1f}")
```
