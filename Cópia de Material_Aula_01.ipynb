## Formatação de Texto

Crie um programa que leia um texto e traga ele em:

1.   Maiuscula
2.   Minuscula
3.   Titulo
4.   Todas as primeiras letras Maiusculas

***Exemplo:***

O texto "texto TEXTO" em maiusculo é TEXTO TEXTO

O texto "texto" em minuscula é texto texto

O texto "texto" em titulo é Texto texto

O texto "texto" em primeira letra maiuscula é Texto Texto





### Teste de Execução


---
| Passo | Texto | Resultado |
| ----- | ---- | --------- |
| 1 | Sol brilha alto | SOL BRILHA ALTO |
|||sol brilha alto |
||| Sol Brilha Alto |
||| Sol brilha alto |
| 2 | Vento sopra suave. | VENTO SOPRA SUAVE.
|||vento sopra suave.|
|||Vento Sopra Suave.|
|||Vento sopra suave. |  
| 3 | Rio murmura baixo. | RIO MURMURA BAIXO. |
|||rio murmura baixo. |
|||Rio Murmura Baixo. |
|||Rio murmura baixo. |  
| 4 | Crianças brincam alegres. | CRIANÇAS BRINCAM ALEGRES.|
|||crianças brincam alegres.|
|||Crianças Brincam Alegres.|
|||Crianças brincam alegres. |  
| 5 | Lua surge noite. | LUA SURGE NOITE.|
|||lua surge noite.|
|||Lua Surge Noite.|
|||Lua surge noite. |  


```python
texto = str(input("Insira seu texto: "))

tipo = int(input("Qual formatação você deseja? \n1. MAIÚSCULA \n2. minúscula \n3. Titúlo \n4. Capitalizar Frase   "))

if(tipo == 1):
# Todas as letras maiúsculas
  fim = texto.upper()
  print(fim)

elif(tipo==2):
# Todas as letras minúsculas
  fim = texto.lower()
  print(fim)

elif(tipo==3):
# Primeira letra de toda palavra maiúscula
  fim = texto.title()
  print(fim)
elif(tipo==4):
# Primeira letra da frase maiúscula
  fim = texto.capitalize()
  print(fim)

else:
  print("Desculpe, não entendi, tente novamente")
```
