## Operadores Comparação

| Símbolo | Operação | Descrição | Utilização |
| --- | --- | --- | --- |
| > | Maior | Retorna verdadeiro (True) se o valor da esquerda for maior do que o da direita | x > y |
| >= | Maior ou Igual | Retorna verdadeiro (True) se o valor da esquerda for maior ou igual ao valor da direita | x > = y |
| < | Menor | Retorna verdadeiro (True) se o valor da esquerda for menor do que o valor da direita | x < y |
| <= | Menor ou Igual | Retorna verdadeiro (True) se o valor da esquerda for menor ou igual ao valor da direita | x < = y |
| == | Igual | Retorna verdadeiro (True) se o valor da esquerda for igual ao valor da direita | x == y |
| != | Diferente | Retorna verdadeiro (True) se o valor da esquerda for diferente do valor da direita | x ! = y |

## Operadores Lógicos

| Símbolo | Descrição | Utilização |
| --- |  --- | --- |
| and | Retorna verdadeiro (True) se e somente se, os dois valores colocados sejam verdadeiros | x and y |
| or | Retorna verdadeiro (True) se qualquer um dos dois seja verdade e retorna falso (False) se e somente se, os dois forem falsos | x or y |
| not | Retorna verdadeiro (True) se o valor for falso (False), retorna falso (False) se o valor for verdadeiro (True) | x not y |

## Maior e Menor

Crie um programa que leia 3 números e me fale qual é o maior entre eles e o menor entre eles

### Teste de Mesa


---

| Teste | número 1 | número 2 | número 3 | resultado |
| - | - | - | - | -|
| 1 | 2487 | 1156 | 3924 | O maior número digitado foi 3924 e o menor número digitado foi 1156 |
| 2 |  789 | 2310 | 4678 | O maior número digitado foi 4678 e o menor número digitado foi 789 |
| 3 | 33 | 1723 | 419 | O maior número digitado foi 1723 e o menor número digitado foi 33 |
| 4 | 2995 | 4210 | 1444 | O maior número digitado foi 4210 e o menor número digitado foi 1444 |
| 5 | 2766 | 871 | 4963 | O maior número digitado foi 4963 e o menor número digitado foi 871 |

```python
n1 = int(input("Digite o primeiro número  "))
n2 = int(input("Digite o segundo número  "))
n3 = int(input("Digite o terceiro número  "))

if(n1 > n2 > n3):
  maior = n1
  menor = n3

elif(n1 > n3 > n2):
  maior = n1
  menor = n2

elif(n2 > n3 > n1):
  maior = n2
  menor = n1

elif(n2 > n1 > n3):
  maior = n2
  menor = n3

elif(n3 > n2 > n1):
  maior = n3
  menor = n1

else:
  maior = n3
  menor = n2

print(f"O maior número digitado foi {maior} e o menor número digitado foi {menor}")
```

## Média

Crie um programa que leia 3 notas, tire a média delas e diga se a pessoa foi aprovada, reprovada ou ficou de recuperação

### Teste de Mesa


---

| Teste | número 1 | número 2 | número 3 | resultado |
| - | - | - | - | -|
| 1 | 4,29 | 8,75 | 2,14 | Recuperação  |
| 2 |  6,98 | 3,51 | 9,23 | Recupeação |
| 3 | 1,87 | 5,62 | 7,43 | Reprovado |
| 4 | 2,56 | 8,01 | 3,79 | Reprovado |
| 5 | 6,14 | 9,87 | 8,15 | Aprovado |

```python
n1 = float(input("Digite a nota 1  ").replace(",","."))
n2 = float(input("Digite a nota 2  ").replace(",","."))
n3 = float(input("Digite a nota 3  ").replace(",","."))
resultado = (n1+n2+n3)/3
print(f"Resultado: {resultado:.1f}")

if resultado >= 7:
  print(f"\033[1;32mAprovado\033[0m")

elif(resultado < 5):
  print(f"\033[4;31mReprovado\033[0m")

else:
  print(f"\033[0;33mRecuperação\033[0m")
```

## Nível de Segurança

Crie um programa onde a pessoa fala o seu nível de acesso entre: administrador ou adm, gerente ou ger, funcionario ou func e caso não seja nenhum desses falar que é desconhecido e não tem acesso a nenhum nível de segurança

### Teste de Mesa


---
| Teste | Nome | Usuario | Resultado |
| ----- | ---- | --------- | -------- |
| 1 | hermione granger | gerente | Olá gerente Hermione Granger você tem acesso ao nível de segurança do funcionário |
| 2 | luke skywalker | func | Olá funcionário Luke Skywalker você tem acesso ao nível de segurança básico |
| 3 | lara croft | adm | Olá adm Lara Croft você tem acesso ao nível de segurança ilimitado |
| 4 | mario bros | encanador | Olá encanador Mario Bros, nosso sistema o classifica como desconhecido, você não tem acesso a nenhum nível de segurança |
| 5 | naruto uzumaki | hokage | Olá hokage Naruto Uzumaki, nosso sistema o classifica como desconhecido, você não tem acesso a nenhum nível de segurança |


```python
nome = str(input("Digite seu nome  ")).title().strip()
user = str(input("Qual sua função?  ")).lower().strip()

if(user == "administrador" or user == "adm"):
  print(f"Olá {user} {nome} você tem acesso ao nível de segurança ilimitado")

elif(user == "gerente" or user == "ger"):
  print(f"Olá gerente {nome} você tem acesso ao nível de segurança do funcionário")

elif(user == "funcionario" or user == "func"):
  print(f"Olá funcionario {nome} você tem acesso ao nível de segurança básico")

else:
  print(f"Olá {user} {nome}, nosso sistema o classifica como desconhecido, você não tem acesso a nenhum nível de segurança")
```

## Pagamento

Crie um programa onde leia o valor da compra e fale para o usuário os tipos de pagamento que oferece:

1. Avista Dinheiro
2. Avista Cartão
3. Avista PIX
4. 2x no Cartão
5. 3x ou mais no Cartão

Após selecionar fale para ele de acordo com o selecionado qual será o valor com desconto ou juros que será dado:

1. 10% de desconto
2. 2% de desconto
3. 25% de desconto
4. 0% de desconto
5. 20% de juros

### Teste de Mesa


---

| Teste | Valor | Tipo de Pagamento | Resultado |
| ----- | ---- | --------- | -------- |
| 1 | 2387,30 | 1 | 2148.57 |
| 2 | 6412,85 | 2 | 6284.59 |
| 3 | 1795,15 | 3 | 1346.36 |
| 4 | 5023,50 | 4 | 5023.50 |
| 5 | 7856,40 | 5 | 9427.68 |

```python
total = float(input("Insira o valor da compra  ").replace(",","."))
pag = int(input(f"Seu total é {total:.2f}, deseja pagar de qual forma?\n1. A vista dinheiro \n2. A vista cartão \n3. A vista PIX \n4. 2x no Cartão \n5. 3x ou mais no cartão  "))

if(pag == 1):
  total *= 0.9


elif(pag == 2):
  total *= 0.98

elif(pag == 3):
  total *= 0.75

elif(pag == 4):
  total *= 1

elif(pag == 5):
  total *= 1.2

else:
  print("Não entendi, tente novamente.")

print(f"O valor final da sua compra é de {total:.2f}")
```

## Triagem Hospital

Crie um programa onde pergunte o nome, idade e se está com suspeita de doença infectocontagiosa.

Caso a pessoa esteja com doença infectocontagiosa ela vai para a sala Amarela e caso ela nao tenha ela vai para a sala Branca.

Depois verifique se a pessoa tem mais de 65 anos, caso ela tenha ela será uma paciente com PRIORIADADE caso não verifque se a pessoa é do genero feminino e se está grávida, se estiver ela será uma paciente com prioridade, caso nenhum dos requisitos seja verdade será um paciente sem prioridade

```python
nome = str(input("Digite seu nome  ")).title().strip()
idade = int(input("Digite sua idade  "))
dinc = str(input("Você está com suspeita de doença infectocontagiosa? (S/N)  ")).strip().upper()

if(dinc == "S"):
  sala = "Amarela"

else:
  sala = "Branca"

if(idade >= 65):
  pac = "PRIORIDADE"

else:
  gen = str(input("Você é do sexo feminino? (S/N)  ")).upper().strip()
  if(gen =="S"):
    grav = str(input("Você está em processo de gravidez? (S/N)  ")).upper().strip()

    if(gen == "S" and grav == "S"):
      pac = "PRIORIDADE"
    else:
      pac = "SEM PRIORIDADE"

  else:
    pac = "SEM PRIORIDADE"

print(f"Por favor, {nome}, encaminhe-se à sala \033[4m{sala}\033[0m, você é um paciente do tipo {pac}")
```
