## Limite

Crie um programa onde a pessoa vai passar um limite e ele terá que mostrar os números de 0 até esse limite

```python
final = int(input("Insira o número final  ").strip())

for num in range(final):
  print(num, end = " ")

print(final)
```

## Tabuada

Crie uma tabuada onde o usuário consiga incluir o número que deseja multiplicar e até qual número ele deseja chegar

```python
num = int(input("Insira o número que deseja multiplicar  ").strip())
limite = int(input("Insira o limite ao qual deseja multiplicar  ").strip())
limite = int(limite/num)
fator = 0

for mult in range(limite+1):
  mult = num * fator
  print(mult, end = " ")
  fator += 1
```

## Adivinhação

Crie um programa onde o computador pense em um número aleatório e a pessoa tenha que tentar advinhar esse número no menor número de tentativas onde o máximo de tentativas é 10

Caso a pessoa erre mostre para ela se ela digitou um número acima ou abaixo do número pensado e depois retorne para o `for`

```python
import random

num = random.randint(0,100)

for tent in range(10):
  chute = abs(int(input("Em que número estou pensando?  ").strip()))

  if(chute > 100):
    print("Não, não tão alto, está entre 0 e 100!")
  elif(chute > num):
    print(f"Não, não é {chute}. O número que pensei é menor")
  elif(chute < num):
    print(f"Não, não é {chute}. O número que pensei é maior")
  elif(chute == num):
    print(f"Isso!", end =" ")
    break


print(f"O número que pensei era {num}")
```

## Soma dos Pares

Crie um programa onde leia 7 números quaisquer e conte quantos números pares apareceram e some eles.

Caso o valor for impar, desconsidere

```python
conta = 0
for i in range(7):
  num = int(input("Insira um número  ").strip())

  if(num % 2 == 0):
    print(f"contagem = {conta} + {num}")
    conta += num

  else:
    continue

print(conta)
```

# Sequência

Crie um programa onde a pessoa digite até qual número ela deseja chegar, Exemplo: 150 e de quantos em quantos números ela deseja pular para chegar neste número ou o mais próximo dele, iniciado do número 0

```python
final = int(input("Insira o valor final ao qual deseja chegar  ").strip())
pulo = abs(int(input("Qual intervalo devo colocar entre os números?  ").strip()))

if(final < 0):
   for i in range(final,1,pulo):
    print(i, end = " ")

else:
   for i in range(0,final+1,pulo):
    print(i, end = " ")
```
