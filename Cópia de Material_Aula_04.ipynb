## Limite

Crie um programa onde a pessoa vai passar um limite e ele terá que mostrar os números de 0 até esse limite

```python
num = abs(int(input("Insira o número final para sua contagem  ")))
i = 0

while i <= num:
  print(i, end=" ")
  i += 1
```

# Media

Crie um programa onde o usuario vai adicionar quantas numeros ele desejar e depois a gente vai tirar a média de todas os numeros que foram incluidos e falar qual foi o maior e qual foi o menor número adicionado

```python
num = float(input("Isira um número   ").strip().replace(",","."))
cont = str(input("Deseja continuar? S/N  ").strip().upper())
valor = 0
i = 1

while cont == "S":
  valor += num
  num = float(input("Isira um número   ").strip().replace(",","."))
  cont = str(input("Deseja continuar? S/N  ").strip().upper())
  i += 1

media = valor/i
print(f"media = {media:.2f}")
```

# Soma de valores

Crie um programa onde leia varios números flutuantes do teclado. O programa só vai parar quando o usuario digitar 0.

No final mostre quantos números foram digitados e qual a soma de todos eles.

```python
num = float(input("Isira um número   ").strip().replace(",","."))
soma = num
i = 1

while num != 0:
  num = float(input("Isira um número   ").strip().replace(",","."))
  soma += num
  i += 1

print(f"Você digitou {i} números, cuja soma é igual a {soma:.2f}")
```

## Calculadora

1 - Somar

2 - Multiplicar

3 - Dividir

4 - Subtrair

5 - Novos números

6 - Sair do Programa

Seu programa deverá realizar a operação solicitada em cada caso

```python
n1 = float(input("Insira o primeiro número  ").strip().replace(",","."))
n2 = float(input("Insira o segundo número  ").strip().replace(",","."))

operacao = abs(int(input('''Qual operação deseja realizar?
1 - Soma
2 - Multiplicação
3 - Divisão
4 - Subtração
5 - Inserir novos números
6 - Sair do programa  ''')))

while operacao == 5:
  n1 = float(input("Insira o primeiro número  ").strip().replace(",","."))
  n2 = float(input("Insira o segundo número  ").strip().replace(",","."))

  operacao = abs(int(input('''Qual operação deseja realizar?
1 - Soma
2 - Multiplicação
3 - Divisão
4 - Subtração
5 - Inserir novos números
6 - Sair do programa  ''')))


if operacao == 1:
  resultado = n1+n2

elif operacao == 2:
  resultado = n1*n2

elif operacao == 3:
  resultado = n1/n2

elif operacao == 4:
  resultado = n1-n2

if operacao != 6:
  print(f"O resultado da sua operação é {resultado}")

```
